# Building a Co-occurrence Matrix from Scratch

```python
import numpy as np
```

```python
# Sample corpus
corpus = [
    ['the', 'cat', 'sat', 'on', 'the', 'mat'],
    ['the', 'dog', 'sat', 'on', 'the', 'mat'],
    ['the', 'cat', 'chased', 'the', 'dog']
]
```

```python
# -------------------------------
# Build Vocabulary
# -------------------------------

vocab = sorted(set(word for sent in corpus for word in sent))
word2id = {word: idx for idx, word in enumerate(vocab)}

n = len(vocab)


```

```python
print("Vocabulary:")
print(vocab)
```

```python
# -------------------------------
# Build Co-occurrence Matrix
# -------------------------------

cooc = np.zeros((n, n))

window = 2

for sentence in corpus:

    for i, word in enumerate(sentence):

        start = max(0, i - window)
        end = min(len(sentence), i + window + 1)

        for j in range(start, end):

            if i != j:

                # Weight = 1 / distance
                distance = abs(i - j)

                cooc[word2id[word]][word2id[sentence[j]]] += 1 / distance
```

```python
# -------------------------------
# Results
# -------------------------------

print("\nCo-occurrence Matrix Shape:")
print(cooc.shape)
```

```python

print("\nCo-occurrence Matrix:")
print(cooc)
```

```python
print("\nCo-occurrence of 'cat' and 'sat':")
print(
    cooc[
        word2id['cat']
    ][
        word2id['sat']
    ]
)
```
